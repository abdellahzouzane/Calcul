```python
import matplotlib.pyplot as plt
import numpy as np

Notes_ue1 = {
    'R101': 10,
    'R102': 10,
    'R103': 7,
    'R104': 7,
    'R106': 5,
    'R108': 6,
    'R110': 5,
    'R111': 4,
    'R112': 2,
    'R113': 5,
    'R114': 5,
    'SAE11': 20,
    'SAE12': 20,
    'SAE16': 7
}
Coef_ue1 = [10, 10, 7, 7, 5, 6, 5, 4, 2, 5, 5, 20, 20, 7]

note_ue1 = Notes_ue1.values()
coefs = Coef_ue1
UE1_ensemble = zip(note_ue1, coefs)

produit1 = []

for note, coef in UE1_ensemble:
    produit1.append(note * coef)
somme_produit1 = sum(produit1)
somme_coefs = sum(coefs)
moyenne_ue1 = somme_produit1 / somme_coefs
print('moyenne de UE1 :', moyenne_ue1)

def get_color(value):
    if value >= 10:
        return 'green'
    elif value >= 8:
        return 'yellow'
    else:
        return 'red'

colors = [get_color(val) for val in [moyenne_ue1]]
plt.bar(["UE1"], [moyenne_ue1], color=colors)
```
